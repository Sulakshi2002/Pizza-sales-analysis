### Daily Trends Total analysis

```python

# 1. Convert 'order_date' safely (handles mixed date formats)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', errors='coerce')

# Drop rows where date couldnt be converted
df = df.dropna(subset=['order_date'])

# 2. Extract day name (Monday, Tuesday, etc.)
df['day_name'] = df['order_date'].dt.day_name()
#This takes the date in each row (e.g., 2025-11-06)and finds which day of the week it is (e.g., Monday, Tuesday, etc.).
#It then stores that name in a new column called 'day_name'.


#3. Define correct weekday order (so chart is in order, not alphabetical)
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['day_name'] = pd.Categorical(df['day_name'], categories=weekday_order, ordered=True)
#This tells Python the correct order of days  Monday to Sunday.
#Normally, when you sort text, Python puts it in AZ order,which would look wrong (Friday, Monday, Saturday...).
#By using Categorical, we make sure the days stay in real calendar order.


# 4. Count unique orders by day of the week
orders_by_day = df.groupby('day_name', observed=False)['order_id'].nunique()
# Groups all rows by their day name (e.g., all Monday orders together).
#'nunique()' counts the number of unique order IDs for each day,so duplicates are not counted twice.
# This gives us how many total orders happened on each weekday.


#5. Plot bar chart
ax = orders_by_day.plot(kind='bar', figsize=(8, 5), color='green', edgecolor='black')

#6. Customize chart
plt.title("Total Orders by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

#7. Add value labels on top of each bar
for i, val in enumerate(orders_by_day):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

#8. Final layout and show chart
plt.tight_layout()
plt.show()
```

### Daily Trends Total revenue

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', errors='coerce')

df = df.dropna(subset=['order_date'])

df['day_name'] = df['order_date'].dt.day_name()

weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['day_name'] = pd.Categorical(df['day_name'], categories=weekday_order, ordered=True)

orders_by_day = df.groupby('day_name', observed=False)['total_price'].sum()

ax = orders_by_day.plot(kind='bar', figsize=(8, 5), color='purple', edgecolor='black')

plt.title("Total Revenue by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

for i, val in enumerate(orders_by_day):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Hourly Trends Total orders

```python
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')

df['order_hour'] = df['order_time'].dt.hour

orders_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique()

ax = orders_by_hour.plot(kind='bar', figsize=(10, 5), color='maroon', edgecolor='black')

plt.title("Total Orders by Hour of Day")
plt.xlabel("Hour Of Day")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

for i, val in enumerate(orders_by_hour):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Monthy Trends Total orders

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', errors='coerce')

df['month_name'] = df['order_date'].dt.month_name()

month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
df['month_name'] = pd.Categorical(df['month_name'], categories=month_order, ordered=True)

orders_by_month = df.groupby('month_name', observed=False)['order_id'].nunique()

#cfreate a line chart
plt.figure(figsize=(12,5))
plt.fill_between(orders_by_month.index, orders_by_month.values, color="orange", alpha=0.6)
plt.plot(orders_by_month.index, orders_by_month.values, color="black", linewidth=2, marker = 'o')

plt.title("Total Order by month")
plt.xlabel("Month")
plt.ylabel("Number of order")
plt.xticks(rotation=45)

for i, val in enumerate(orders_by_month):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

### % of Sales by Category

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()

category_pct = category_sales/category_sales.sum() * 100

plt.figure(figsize=(5,5))
colors = plt.get_cmap('tab20').colors  #nice color palette

plt.pie(category_pct, labels=category_pct.index, autopct = '%1.1f%%', startangle=90, colors = colors, wedgeprops={'edgecolor':'black','width':0.4})
plt.title("Precentage sales by pizza category")
plt.show()
```

###  % Sales by Pizza category and size

```python
sales_pivot = df.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0
)
sales_pct = sales_pivot / sales_pivot.sum().sum()*100

plt.figure(figsize=(10,6))
sns.heatmap(sales_pct, annot=True, fmt=".1f", cmap="YlOrRd", linewidth=0.5)
plt.title("% of Sales by Pizza Category and Size")
plt.ylabel("Pizza category")
plt.xlabel("Pizza size")
plt.show()
```
