## KPI's

```python
total_revenue = df['total_price'].sum()
total_pizzas_sold = df['quantity'].sum()
total_orders = df['order_id'].nunique() #returns the number of unique order ID
avg_order_value= total_revenue / total_orders
avg_pizzas_per_order = total_pizzas_sold / total_orders

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total pizzas sold: {total_pizzas_sold:,}")
print(f"Total orders: {total_orders:,}")
print(f"Average order value: ${avg_order_value:,.2f}")
print(f"Average pizza per order: {avg_pizzas_per_order:,.2f}")
```

### Ingredient analysis

```python
ingredient = (
    df['pizza_ingredients']
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .reset_index()
    .rename(columns = {'index' : 'counts',
                     'pizza_ingredients' : 'Ingredients'})
)
print(ingredient.head(15))
```

### Daily Trends Total analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

#  1. Convert 'order_date' safely (handles mixed date formats)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', errors='coerce')

# Drop rows where date couldnt be converted
df = df.dropna(subset=['order_date'])

#  2. Extract day name (Monday, Tuesday, etc.)
df['day_name'] = df['order_date'].dt.day_name()

#  3. Define correct weekday order (so chart is in order, not alphabetical)
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['day_name'] = pd.Categorical(df['day_name'], categories=weekday_order, ordered=True)

#  4. Count unique orders by day of the week
orders_by_day = df.groupby('day_name', observed=False)['order_id'].nunique()

#  5. Plot bar chart
ax = orders_by_day.plot(kind='bar', figsize=(8, 5), color='green', edgecolor='black')

#  6. Customize chart
plt.title("Total Orders by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

#  7. Add value labels on top of each bar
for i, val in enumerate(orders_by_day):
    plt.text(i, val + 2, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

#  8. Final layout and show chart
plt.tight_layout()
plt.show()
```
